```python
df = pd.read_csv('/content/gdrive/MyDrive/gold_eda.csv')
```

```python
df=df.iloc[:,1:]
```

```python
df
```

```python
df['GOLDAM_Trf'] = np.log(df['gold_rate'])
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['GOLDAM_Trf'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
if (result[1]<0.05):
    print('Data is stationary')
else:
    print('Data is not stationary')
```

```python
df['GOLDAM_srt'] = np.sqrt(df['gold_rate'])
```

```python
result = adfuller(df['GOLDAM_srt'].dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
if (result[1]<0.05):
    print('Data is stationary')
else:
    print('Data is not stationary')
```

```python
df['GOLDAM_diff'] = df['gold_rate'] - df['gold_rate'].shift(1)
```

```python
result = adfuller(df['GOLDAM_diff'].dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
if (result[1]<0.05):
    print('Data is stationary')
else:
    print('Data is not stationary')
```
